```python
from requests import get
from bs4 import BeautifulSoup
from time import sleep
```

```python
base_url1 = "https://quotes.toscrape.com/page/1/"

response = get(base_url1) 

soup = BeautifulSoup(
    response.content, 
    'html.parser'
    )
```

```python
quotes = soup.find_all('div', attrs={'class':'quote'})

```

```python
for index, quote in enumerate(quotes):

    with open(f'quotes/quotes_{index}.txt', 'w') as file:
        file.write(quote.text)



```

```python
base_url = "https://quotes.toscrape.com/"

for page in range(1, 10):
    # Seiten-URL zusammensetzen
    url = base_url + str(page)
    print(url)

    # Seite abfragen
    response = get(url)

    # Suchkartei erstellen
    soup = BeautifulSoup(
        response.content,
        "html.parser"
    )

    quotes = soup.find_all('div', attrs={'class':'quote'})
    for index, quote in enumerate(quotes):

        with open(f'quotes/quote_{page}_{index}.txt', 'w') as file:
            file.write(quote.text)

    # Pause
    sleep(1)


```
