# Gathering Text on the Web

```python
from requests import get
from bs4 import BeautifulSoup
```

## Requesting Web-Content via HTTP

```python
base_url = "https://quotes.toscrape.com/"

response = get(base_url)

soup = BeautifulSoup(
    response.content,
    "html.parser"
)
```

## Parsing Web-Content in Python

```python
quotes = soup.find_all(
    "div",
    attrs={"class": "quote"}
)
```

```python
for quote in quotes:
    print(quote.text)
```
